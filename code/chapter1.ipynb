```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# 1. Load the dataset from the CSV file
# Ensure 'performance_25_final.csv' is in the same directory as this script
data_path = Path('doc/Performance Data .csv')
if not data_path.exists():
    data_path = Path('..') / data_path
df = pd.read_csv(data_path)

# Extract the target column
performance_data = df['Performance Index']

# ==========================================
# 3.1 GRAPHS
# ==========================================
plt.figure(figsize=(12, 10))

# Histogram
# plt.subplot(1, 2, 1)
sns.histplot(performance_data, bins=7, kde=True, color='skyblue')
plt.title('Histogram of Performance Index')
plt.xlabel('Performance Index')
plt.ylabel('Frequency')


plt.tight_layout()
plt.savefig('histogram_PI.png', dpi=200)
plt.show()
```

```python
# Boxplot
# plt.subplot(1, 2, 2)
sns.boxplot(x=performance_data, color='lightgreen')
plt.title('Boxplot of Performance Index')
plt.xlabel('Performance Index')

plt.tight_layout()
plt.savefig('boxplot_PI.png', dpi=200)
plt.show()
```

```python
# ==========================================
# 3.2 SUMMARY STATISTICS
# ==========================================
# These results are used to verify your mandatory hand calculations
mean_val = performance_data.mean()
median_val = performance_data.median()
mode_val = performance_data.mode()[0]
range_val = performance_data.max() - performance_data.min()
variance_val = performance_data.var()  # Sample variance (n-1)

std_dev_val = performance_data.std()
q1 = performance_data.quantile(0.25)
q3 = performance_data.quantile(0.75)
iqr_val = q3 - q1

print("--- Chapter 1: Descriptive Statistics Results ---")
print(f"Mean (x): {mean_val:.2f}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
print(f"Range: {range_val}")
print(f"Sample Variance (s): {variance_val:.2f}")
print(f"Sample Standard Deviation (s): {std_dev_val:.2f}")
print(f"Interquartile Range (IQR): {iqr_val}")
```
