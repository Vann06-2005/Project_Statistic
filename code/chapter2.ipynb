```python
import pandas as pd
import numpy as np
from pathlib import Path

# Load your actual 25 observations from the uploaded file
data_path = Path('Student_Perform.csv')
if not data_path.exists():
    data_path = Path('..') / data_path
df = pd.read_csv(data_path)
data = df['Performance Index']

# Calculations for Chapter 2
n = len(data)  # This will be 25
mean_val = data.mean()
variance_val = data.var()
std_dev_val = data.std()
se_val = std_dev_val / np.sqrt(n)

print(f"--- Chapter 2: Point Estimation Results ---")
print(f"Sample Mean (x̄): {mean_val:.4f}")
print(f"Sample Variance (s²): {variance_val:.4f}")
print(f"Sample Standard Deviation (s): {std_dev_val:.4f}")
print(f"Standard Error (SE): {se_val:.4f}")
```
